```python
import torch
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("lvwerra/distilbert-imdb")
model = AutoModelForSequenceClassification.from_pretrained("lvwerra/distilbert-imdb")
```

```python
inputs = tokenizer("I love using Hugging Face Transformers!", return_tensors="pt").to("cpu")

with torch.no_grad():
    outputs = model(**inputs)

predicted_class_id = torch.argmax(outputs.logits, dim=-1).item()
predicted_label = model.config.id2label[predicted_class_id]
print(f"Predicted label: {predicted_label}")
```

# now let's feed the model one token at a time

```python
from datasets import load_dataset

ds = load_dataset("stanfordnlp/imdb")
```

```python
test_phrase = ds['train'][0]['text']
# print(type(test_phrase))
# print(test_phrase)

phrase_tokens = tokenizer.tokenize(test_phrase)
# print(phrase_tokens)

input_ids = []

for word in phrase_tokens:
    # Encode single token and convert to tensor
    new_input = tokenizer.encode(word, return_tensors='pt', add_special_tokens=False)
    input_ids.append(new_input)

    # Concatenate input so far
    full_input = torch.cat(input_ids, dim=-1)

    # Run the model
    with torch.no_grad():
        outputs = model(full_input)

    print(f"Token added: {word}")
    print(f"Logits shape: {outputs.logits.shape}")  # [1, seq_len, vocab_size]
```
